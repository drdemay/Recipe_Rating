```python
# import dependencies
# import warnings
# warnings.simplefilter('ignore')

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as seabornInstance 


import sqlite3
from sqlite3 import Error

import sklearn

from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
# Read the csv file into a pandas DataFrame
ratings = pd.read_csv('Resources/epicurious_ratings.csv')
ratings
```

```python
# Create new dataframe with NaN values removed
ratings = ratings.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
# ratings
```

```python
# Reset the index on the new dataframe
ratings.reset_index(drop=True)
```

```python
# delete all rows with column 'calories' has value > 2000
ratings_remove_high_cals = ratings[ (ratings["calories"] >= 2000.00) ].index
ratings.drop(ratings_remove_high_cals, inplace=True)
# ratings
```

```python
# delete all rows with column 'calories' has value 0 to 100
remove_low_cals = ratings[ (ratings["calories"] >= 0) & (ratings['calories'] <= 100) ].index
ratings.drop(remove_low_cals , inplace=True)
# ratings
```

```python
# delete all rows with column 'sodium' has value >= 10000
remove_high_sodium = ratings[ (ratings["sodium"] >= 10000.00) ].index
ratings.drop(remove_high_sodium, inplace=True)
ratings
```

```python
# write this back to csv and for upload to gitHub for Tableau analysis
# ratings.to_csv(r"C:\Users\Deanna\Desktop\Project3\Recipe_Rating\Resources\epi_ratings_clean.csv")  
```

```python
# remove all but the columns that will be using for machine learning
keep_cols = ["calories", "protein", "fat", "sodium", "rating"]
# Potential side chart / analysis - speciality diet ratings / specific ingrediants include: "vegan", "vegetarian", "dairy free", "wheat/gluten-free", 
# "peanut free", "meat", "alcoholic", "chocolate", "cheese"
```

```python
# Create updated dataframe to be used for ML
ratings_nutrition = ratings[keep_cols]
ratings_nutrition
```

```python
# Reset the index on the new dataframe
ratings_nutrition.reset_index(drop=True)
```

```python
# write this back to csv and for upload to gitHub
ratings_nutrition.to_csv(r"C:\Users\Deanna\Desktop\Project3\Recipe_Rating\Resources\epi_ratings_columns_removed.csv")  
```

```python
# Check for linearity

# Single linear regression
# None of them (calories, fat, sodium, protein) are really showing anything
ratings_nutrition.plot(x='rating', y="calories", style='o')  
plt.title('Rating vs Calories')  
plt.xlabel('Rating')  
plt.ylabel('Calories')  
plt.show()
```

We do not have linear data.   
We will create a neural network using Keras.

```python
# Addtional cleaning of data, drop the ratings column of information for training of our model
data = ratings_nutrition.drop("rating", axis=1)
feature_names = data.columns
data
```

```python
# Do a grouping / distribution of ratings
rating_group=ratings.groupby("rating")
print(rating_group["calories"].count())
```

## Applying One-Hot Encoding

```python
# Step 0: Reformat data
data = ratings_nutrition.values
X = data[:, 0:4]
y = data[:, 4]
```

```python
from sklearn.preprocessing import LabelEncoder

# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
```

```python
for label, original_class in zip(encoded_y, y):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```

```python
from tensorflow.keras.utils import to_categorical

# Step 2: One-hot encoding
one_hot_y = to_categorical(encoded_y)
one_hot_y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, one_hot_y, random_state=42)
y_test
```

```python
one_hot_y[0]
```

```python
# Scale, Normalize, and Transform dataframe
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler model and fit it to the training data
X_scaler = StandardScaler().fit(X_train)
```

```python
# Transform the training and testing data using the X_scaler
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```

## Normal Neural Network

```python
# first, create a normal neural network with 4 inputs, 12 hidden nodes, and 8 outputs
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(units=12, activation='relu', input_dim=4))
model.add(Dense(units=8, activation='softmax'))
```

```python
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam',
             loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Fit the model to the training data
model.fit(
    X_train_scaled,
    y_train,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

## Deep Learning

For this network, we are adding in additional layers

```python
deep_model = Sequential()
deep_model.add(Dense(units=6, activation='relu', input_dim=4))
deep_model.add(Dense(units=12, activation='relu'))
deep_model.add(Dense(units=12, activation='relu'))
deep_model.add(Dense(units=8, activation='softmax'))
```

```python

deep_model.summary()
```

```python
deep_model.compile(optimizer='adam',
                   loss='categorical_crossentropy',
                   metrics=['accuracy'])

deep_model.fit(
    X_train_scaled,
    y_train,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

## Compare the models below

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
model_loss, model_accuracy = deep_model.evaluate(
    X_test_scaled, y_test, verbose=2)
print(f"Deep Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

# Combine ratings to create less outputs

2.500, 3.125 = 2.0   
3.750 = 3.0  
4.375 = 4.0  
5.000 = 5.0

```python
# Use cleaned dataframe
ratings_nutrition
```

```python
# for loop to assign new ratings
new_rating = [] 
for value in ratings_nutrition["rating"]: 
    if value <= 2.5: 
        new_rating.append(1) 
    elif value == 3.125: 
        new_rating.append(2) 
    elif value == 3.75: 
        new_rating.append(3) 
    elif value == 4.375: 
        new_rating.append(4)
    elif value == 5.000: 
        new_rating.append(5)
    else: 
        new_rating.append("Error") 
       
ratings_nutrition["new rating"] = new_rating    

# # for loop to assign new ratings
# new_rating = [] 
# for value in ratings_nutrition["rating"]: 
#     if value <= 1.875: 
#         new_rating.append(1) 
#     elif value >= 2.50 and value <= 3.125: 
#         new_rating.append(2) 
#     elif value == 3.75: 
#         new_rating.append(3) 
#     elif value == 4.375: 
#         new_rating.append(4)
#     elif value == 5.000: 
#         new_rating.append(5)
#     else: 
#         new_rating.append("Error") 
       
# ratings_nutrition["new rating"] = new_rating    
```
